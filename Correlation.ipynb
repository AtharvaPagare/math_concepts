```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('toy_datasets\Wine.csv', header = None, names = ['Wine', 'Alcohol', 'Malic.acid', 'Ash', 'Acl', 'Mg', 'Phenols', 'Flavanoids', 'Nonflavanoid.phenols', 'Proanth', 'Color.int', 'Hue', 'OD', 'Proline'])
```

```python
#names = ['Wine', 'Alcohol', 'Malic.acid', 'Ash', 'Acl', 'Mg', 'Phenols', 'Flavanoids', 'Nonflavanoid.phenols', 'Proanth', 'Color.int', 'Hue', 'OD', 'Proline']
#l = len(names)
X = data.drop('Wine', axis = 1)
Y = data['Wine']
```

```python
X.head()
print(len(X.columns))
```

```python
Y.head()
```

```python
from scipy.stats import pearsonr
```

```python
X.Alcohol
```

```python
from scipy.stats import spearmanr
thresh = 0.5
coef_ = []
new_feat = []
i = -1
feat_names = ['Alcohol', 'Malic.acid', 'Ash', 'Acl', 'Mg', 'Phenols', 'Flavanoids', 'Nonflavanoid.phenols', 'Proanth', 'Color.int', 'Hue', 'OD', 'Proline']
for name in feat_names:
    coef_.append(spearmanr(X[name],Y)[0])
    i += 1
    if(coef_[i]>thresh or coef_[i]<-thresh):
        new_feat.append(name)
#print(coef_)
```

```python
coef_
```

```python
new_feat
```

```python
X_new_feat = X[['Acl', 'Phenols', 'Flavanoids', 'Proanth', 'Hue', 'OD', 'Proline']]
```

```python
X_new_feat
```

```python
from sklearn.model_selection import train_test_split
```

```python
#analysis including all features
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2, random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
print('Accuracy score while considering all the features is ',':%0.3f'%(accuracy_score(pred, y_test)))
```

```python
#analysis including reduced features with high pearson coefficient of correaltion
X_new_feat_train, X_new_feat_test, y_train, y_test = train_test_split(X_new_feat,Y,test_size = 0.2, random_state = 10)
```

```python
clf_new_feat = RandomForestClassifier()
```

```python
clf_new_feat.fit(X_new_feat_train, y_train)
```

```python
pred_new_feat = clf_new_feat.predict(X_new_feat_test)
```

```python
print('Accuracy score while considering all the features is ',':%0.3f'%(accuracy_score(pred_new_feat, y_test)))
```
